```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# Importing the development dataset

dataset = pd.read_csv('Basic Data.csv')
x = dataset.iloc[:, 2:-1]
y = dataset.iloc[:, -1]
```

```python
# Encoding the y values ( 0 -> negative, 1 -> positive)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Splitting the dataset into traning(75) and test(25) sets

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25,
                                                    random_state = 0)
```

```python
# Replacing missing values of training data 

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy='mean')
imputer.fit(x_train)
x_train = imputer.transform(x_train)
```

```python
# Replacing missing values of testing data

imputer.fit(x_test)
x_test = imputer.transform(x_test)
```

```python
# Feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test) 
# transforming the test set with the same scale as the training set
```
